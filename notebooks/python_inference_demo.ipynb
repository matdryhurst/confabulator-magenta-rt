```python
import IPython.display as ipd

from magenta_rt import MagentaRT2Jax
from magenta_rt import MagentaRT2Mlx
from magenta_rt import MagentaRT2Mlxfn
```

## Jax API

```python
mrt = MagentaRT2Jax(size="mrt2_base")
embedding = mrt.embed_style("disco funk")
```

```python
wav, state = mrt.generate(state=None, style=embedding, notes=None, frames=25)
```

```python
ipd.display(ipd.Audio(wav.samples.T, rate=48000))
```
